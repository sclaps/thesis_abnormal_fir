```python
from google.cloud import bigquery

# Initialize client
# client = client = bigquery.Client()  # Uncomment this line if you haven't initialized the client earlier

# Write your SQL query
query = """
    SELECT 
        FlightLegUtcId,
        AircraftType,
        AircraftSubType,
        AircraftRegistration,
        DepartureAirport,
        ActualArrivalAirport,
        SOBT,
        AOBT,
        ATOT,
        ATDT,
        AIBT
    FROM `afkl-skyarch-p200.ds_refined_mart_atm.KlmFlightLegs`
    WHERE 1=1
        AND FlightLegUtcId >= '2026-03-01'
    ORDER BY SDD DESC, FlightLegUtcId ASC
"""

# Execute the query
print("Running query...\n")
res_df = client.query(query).to_dataframe()

# Get and display results
display(res_df)
```
